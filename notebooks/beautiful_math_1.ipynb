```python
import numpy as np
import matplotlib.pyplot as plt

# Define the grid
x = np.linspace(-1, 1, 400)
y = np.linspace(-1, 1, 400)
X, Y = np.meshgrid(x, y)

# Define the complex plane
Z = X + 1j * Y

# Define the Möbius transformation
A = 1
B = 2
C = 3
D = 4
Z_transformed = (A * Z + B) / (C * Z + D)

# Plot the original grid
plt.subplot(1, 2, 1)
plt.imshow(np.angle(Z), extent=(-1, 1, -1, 1), origin='lower', cmap='hsv')
plt.title('Original')

# Plot the transformed grid
plt.subplot(1, 2, 2)
plt.imshow(np.angle(Z_transformed), extent=(-1, 1, -1, 1), origin='lower', cmap='hsv')
plt.title('Transformed')

plt.show()
```
